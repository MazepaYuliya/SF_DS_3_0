## Комбинаторика

| | Перестановки | Размещения | Сочетания |
|--|--|--|--|
| Без повторений |	$P_n = n!$ | $A_n^k = \cfrac{n!}{(n-k)!}$ |	$C_n^k = \cfrac{n!}{(n-k)!k!}$	|
| С повторениями |	$P_n = \cfrac{n!}{k_1!k_2!...k_n!}$ | $\overline{A_n^m} = n^m$	| $\overline{C_n^m} = C_{n+m-1}^m = \cfrac{(n+m-1)!}{(n-1)!m!}$ |		

### Примеры

**Перестановки без повторений**:

Cколько возможных перестановок четырёхзначных чисел можно составить из четырёх разных цифр?

На 1-м месте 4 варианта цифр, на втором 3 и т.д. Итого $P_4 = 4! = 24$

**Перестановки с повторениями**:

Найти количество последовательностей, переставляя буквы в слове «параллелограмм».

Если бы все буквы были уникальными, количество перестановок было бы равно: $P_5 = 5! = 120$

Однако в слове три раза повторяется буква «а», два раза — буква «р», три раза — буква «л» и два раза — буква «м». Вне зависимости от того, поставим мы в начале первую «м» или вторую, слова для нас будут выглядеть одинаково. Поэтому мы должны поделить количество перестановок на количество перестановок каждой конкретной буквы:

$\cfrac{14!}{2!2!3!3!} = 605404800$

**Размещения без повторений**:

Допустим, в комнату вошли десять человек. Мы хотим посадить их на три стула, и порядок рассадки важен.

$A_{10}^3 = \cfrac{10!}{10-3!} = 10*9*8 = 720$

**Размещения с повторениями**:

Мы хотим составить комбинацию из трёх цифр, причём каждая цифра может использоваться любое количество раз.

$\overline{A_{10}^3} = 10^3 = 1000$

**Сочетания без повторений**:

Предположим, в компании работают три женщины и четверо мужчин. Мы хотим выбрать две женщины и двоих мужчин для поездки в командировку. Сколькими способами мы можем это сделать?

$C_3^2 * С_4^2 = \cfrac{3!}{(3-2)!2!} * \cfrac{4!}{(4-2)!2!} = 3 * 6 = 18$

**Сочетания с повторениями**:

Предположим, мы хотим найти количество способов, которыми можно выбрать пять фруктов из корзины, в которой всего 12 фруктов. Мы можем возвращать фрукты и выбирать их заново.

$\overline{C_{12}^5} = C_{12+5-1}^5 = \cfrac{16!}{11!5!} = 4368$

### Бином Ньютона

$(x + y)^n = \sum_{k=0}^n C_n^kx^ky^{n-k}$

Например, 

$(x - 2y)^6 = \\ C_6^0*x^6 - C_6^1*x^5*2y + C_6^2*x^4*4y^2 - C_6^3*x^3*8y^3 + C_6^4*x^2*16y^4 - C_6^5*x*32y^5 + C_6^6*64y^6 = \\ x^6 - 12x^5y + 60x^4y^2 - 160x^3y^3 + 240x^2y^4 - 192xy^5 + 64y^6$
