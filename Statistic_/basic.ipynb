## Основы статистики

Способы формирования выборки:
* Простая случайная выборка;
* Стратифицированная выборка - разбиение генеральной совокупности на страты и случайная выборка из каждой страты (например, разбиение на мужчин и женщин);
* Групповая выборка (кластерная) - разбиение на похожие кластеры (например, регионы города) и выборка из нескольких произвольных кластеров;

Типы переменных:
* Количественные:
    - непрерывные;
    - дискретные;
* Качественные/категориальные:
    - номинативные;
    - ранговые;

Меры центральной тенденции:
* мода;
* медиана;
* среднее;

Свойства среднего:
* $M_{x+c} = M_x + c$ - если к каждому значению выборки прибавить число, то и среднее изменится на это число;
* $M_{x*c} = M_x * c$
* $\sum(x_i - M) = 0$

Меры изменчивости:
* размах $x_{max} - x_{min}$;
* дисперсия:
    - для генеральной совокупности $D = \cfrac{\sum (x_i - \=x)^2}{n}$;
    - для выборки $D = \cfrac{\sum (x_i - \=x)^2}{n-1}$;
* среднее квадратическое отклонение:
    - для генеральной совокупности $\sigma = \sqrt{D}$;
    - для выборки стандартное отклонение $Sd = \sqrt{D}$

Свойства дисперсии:
* $D_{x+c} = D_x$ - если к каждому значению выборки прибавить число, то дисперсия не изменится;
* $sd_{x+c} = sd_x$ - если к каждому значению выборки прибавить число, то дисперсия не изменится;
* $D_{x*c} = D_x * c^2$
* $sd_{x*c} = sd_x * c$

Нормальное распределение:
* унимодально;
* симметрично;
* отклонение наблюдений от среднего подчиняется определенному вероятностному закону;



Стандартизация (z-преобразование) - преобразование полученных данных в стандартную Z-шкалу (Z-scores) c $M_z = 0$ и $D_z = 1$

$Z_i = \cfrac{x_i - \=x}{\sigma_x}$

* $M \pm \sigma \approx 68\%$ наблюдений
* $M \pm 2*\sigma \approx 95\%$ наблюдений
* $M \pm 3*\sigma \approx 100\%$ наблюдений

Задача:
Данные распределены по нормальному закону. Среднее по выборке = 150, стандартное отклонение = 8. Какой % наблюдений первосходит 154.

Решение:
$Z_{154} = \cfrac{x_i - \=x}{\sigma_x} = \cfrac{154 - 150}{8} = 0.5$
